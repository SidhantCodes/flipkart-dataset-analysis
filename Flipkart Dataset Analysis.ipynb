# Explanatory Data Analysis

We shall be analysing our data based on the parameters that what product was most frequently brought, which gender outnumbered the other in spending money on flipkart

On the basis of gender:

```python
ax = sns.countplot(x = 'gender',data = df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales = df.groupby(['gender'], as_index=False)['actprice1'].sum().sort_values(by='actprice1', ascending=False)
sns.barplot(x = 'gender',y= 'actprice1' ,data = sales)
```

```python
sales = df.groupby(['gender'], as_index=False)['Rating'].sum().sort_values(by='Rating', ascending=False)
sns.barplot(x = 'gender',y= 'Rating' ,data = sales)
```

```python
sales = df.groupby(['gender'], as_index=False)['norating1'].sum().sort_values(by='norating1', ascending=False)
sns.barplot(x = 'gender',y= 'norating1' ,data = sales)
```

```python
sales = df.groupby(['gender'], as_index=False)['noreviews1'].sum().sort_values(by='noreviews1', ascending=False)
sns.barplot(x = 'gender',y= 'noreviews1' ,data = sales)
```

```python
sales = df.groupby(['gender'], as_index=False)['star_5f'].sum().sort_values(by='star_5f', ascending=False)
sns.barplot(x = 'gender',y= 'star_5f' ,data = sales)
```

```python
sales = df.groupby(['gender'], as_index=False)['star_4f'].sum().sort_values(by='star_4f', ascending=False)
sns.barplot(x = 'gender',y= 'star_4f' ,data = sales)
```

```python
sales = df.groupby(['gender'], as_index=False)['star_3f'].sum().sort_values(by='star_3f', ascending=False)
sns.barplot(x = 'gender',y= 'star_3f' ,data = sales)
```

```python
sales = df.groupby(['gender'], as_index=False)['star_2f'].sum().sort_values(by='star_2f', ascending=False)
sns.barplot(x = 'gender',y= 'star_2f' ,data = sales)
```

```python
sales = df.groupby(['gender'], as_index=False)['star_1f'].sum().sort_values(by='star_1f', ascending=False)
sns.barplot(x = 'gender',y= 'star_1f' ,data = sales)
```

From the above graphs obtained we can draw the following conclusions:

1. Women outnumber men in terms of shopping online from flipkart

2. Women again come on top in terms of spending money when it comes to buying products online

3. The number of women who have given ratings to products> men

4. However although (3) holds true, yet when indivisually each level of rating was compared it was observed that men had always given more rating than women did in a particular category.


# Analysing the products


```python
#Viewing the top 10 rated products that were sold
sorted_df = df.sort_values(by='Rating', ascending=False)

top_ten_df = sorted_df.head(10)

plt.figure(figsize=(10, 6))
sns.barplot(x='title', y='Rating', data=top_ten_df)
plt.title("Top Ten Rated Products")
plt.xlabel("Product Name")
plt.ylabel("Rating")
plt.xticks(rotation=45)
plt.show()
```

```python
sorted_df = df.sort_values(by='star_5f', ascending=False)

top_ten_df = sorted_df.head(10)

plt.figure(figsize=(10, 6))
sns.barplot(x='title', y='star_5f', data=top_ten_df)
plt.xlabel("Product Name")
plt.ylabel("5 Star Rating")
plt.xticks(rotation=45)
plt.show()
```

```python
sorted_df = df.sort_values(by='star_4f', ascending=False)

top_ten_df = sorted_df.head(10)

plt.figure(figsize=(10, 6))
sns.barplot(x='title', y='star_4f', data=top_ten_df)
plt.xlabel("Product Name")
plt.ylabel("4 Star Rating")
plt.xticks(rotation=45)
plt.show()
```

```python
sorted_df = df.sort_values(by='price1', ascending=False)

top_ten_df = sorted_df.head(10)

plt.figure(figsize=(10, 6))
sns.barplot(x='title', y='price1', data=top_ten_df)
plt.xlabel("Product Name")
plt.ylabel("price")
plt.xticks(rotation=45)
plt.show()
```

From the above few graphs obtained we can conclude:
1. Among the top rated products sold(5 start) there were mainly slippers and footwears
2. The most expensive product that was sold as well belonged to footwear category
